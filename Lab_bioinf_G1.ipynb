```python
def busqueda(secuencia, problema):
  encontrados = 0 #contador para saber cuantas veces se encuentra el problema
  problema = problema #asigna el problema asignado a la variable local
  ubicacion = [] #arreglo para guardar las posiciones

  for i in range(len(secuencia)): #for para hacer el recorrido por el genoma
    if secuence[i:i+len(problema)] == problema: #comparacion del problema asignado con los elementos del genoma
      encontrados += 1 #guarda el registro de las coincidencias encontradas
      ubicacion.append(i) #añade el valor de la ubicacion coincidente al arreglo de posiciones

  print("Numero de veces que se encontro el problema asignado: ", encontrados)
  print("posicion inicial del problema asignado", ubicacion)
  a=(ubicacion[0]) #guarda el valor numerico de la ubicacion inicial
  for i in range(len(ubicacion)): #for para recorrerel arreglo de ubicaciones
    ubicacion[i] += len(problema)-1 #suma la longitud del problema asignado a la ubicacion original con el fin de encontrar la posision final
  b=(ubicacion[0])  #guarda el valor numerico de la ubicacion final
  print("Posicion final", ubicacion)
  pos_in_fi=(a,b)#tupla con los valores inicial y final
  print(pos_in_fi)#imprime la posición inicial y final, en una tupla
```
